```python
list_data = []
folder = '/home/mike-pi/Documents/coding/projects/disp/data/dikson'


for file in os.listdir(folder):
    
    with open(os.path.join(folder, file), 'r') as f:
        text = f.read()
    
    text_=text.split('\n\n')[0]
    date = ' '.join(text_.split(sep='\n')[-1].split()[2:5])

    if 'Дисп' in text:
        # print('ok')
        data = {}
        text = text.strip().split(sep='\n\n')
    
    # print(text)

        for sentence in text:
            if sentence == ' ':
                text.pop(text.index(sentence))
            

        for i in text:
            if i != '':
                if i.split()[0].isdigit() and len(i.split()[0]) <= 3:
                    data.update({'date': date})
                    data.update({i.split()[0]: ' '.join(i.split()[1:])})
            else:
                pass
    else:
        continue        
    list_data.append(data)            
```

```python
for n in list_data:
    print(n)
```

```python
df = pd.DataFrame(list_data).sort_values(by='date').drop(columns='1')
```
